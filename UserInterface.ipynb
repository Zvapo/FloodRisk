```python
import pandas as pd
import numpy as np
import flood_tool
import webbrowser
import os

def get_postcode():
    print("For multiple postcode please seperate by ', '  (e.g TN6 3AW, BN7 2HP, BN1 5PF)")
    postcodes = input ("Postcode: ")
    POSTCODES = postcodes.split(", ")
    return POSTCODES

def get_OSGB36_locations():
    print("For multiple Easting please seperate by ', ' (e.g 530401.0, 541934.0)")
    Easting = input ("Easting: ")
    print("\nFor multiple Northing please seperate by ', ' (e.g 105619.0, 110957.0)")
    Northing = input ("Northing: ")
    Easting = Easting.split(", ")
    Northing = Northing.split(", ")
    
    East = []
    North = []
    for i in Easting:
        East.append(float(i))
    for i in Northing:
        North.append(float(i))
        
    location = [East,North]
    return location

def get_WGS84_locations():
    print("For multiple location please seperate by ', ' (e.g 55.5, -1.54  )")
    Latitude = input ("Latitude: ")
    Longitude = input ("Longitude: ")
    Latitude = Latitude.split(", ")
    Longitude = Longitude.split(", ")
    
    Lat = []
    Long = []
    for i in Latitude:
        Lat.append(float(i))
    for i in Longitude:
        Long.append(float(i))
        
    location = [Lat,Long]
    
    return location

def UserInterface():
    new_tool = flood_tool.Tool(sample_labels=['riskLabel','medianPrice'])
    methods = new_tool.get_default_methods()
    new_tool.train(method=methods)
    
    
    tool = flood_tool.Tool()
    unlabelled = tool.postcodedb
    
    print('\033[1m' +"           FLOOD RISK PREDICTION SOFTWARE BY TEAM TRENT\n")
    print ('\033[0m'+"Available Option\n" 
           "0. Exit\n"
           "1. Predict Flood Class from Postcode\n"
           "2. Predict Flood Class from OSGB36 Location\n"
           "3. Predict Flood Class from WGS84 Location\n"
           "4. Get OSGB36 easthing and northing from Postcode\n"
           "5. Get GPS latitude and longitude from Postcode\n"
           "6. Estimate Local Authority\n"
           "7. Estimate Median House Price\n"
           "8. Estimate Total Property Value\n"
           "9. Estimate Total Annual Flood Risk\n"
           "99. Help!\n"
          )
            
    x = input("Your Choice :")

    if int(x) == 0:
        return print('\033[1m' +'Good Bye')
    elif int(x) == 1: #1. Predict Flood Class from Postcode
        POSTCODES = get_postcode()
        result = new_tool.get_flood_class_from_postcodes(POSTCODES,method=methods['riskLabel'])
        
    elif int(x) == 2: #Predict Flood Class from OSGB36 Location
        coords = get_OSGB36_locations()
        result = new_tool.get_flood_class_from_OSGB36_locations(coords[0],coords[1],method=methods['location'])

    elif int(x) == 3: #Predict Flood Class from WGS84 Location
        coords = get_WGS84_locations()
        result = new_tool.get_flood_class_from_WGS84_locations(coords[0],coords[1],method=methods['location'])
        
    elif int(x) == 4: #Get OSGB36 easthing and northing from Postcode
        POSTCODES = get_postcode()
        result = new_tool.get_easting_northing(POSTCODES)

    elif int(x) == 5: #Get GPS latitude and longitude from Postcode
        POSTCODES = get_postcode()
        result = new_tool.get_lat_long(POSTCODES)
        
    elif int(x) == 6: #Get Local Authority
        coords = get_OSGB36_locations()
        result = new_tool.get_local_authority_estimate(coords[0],coords[1],method=methods['location'])
        
    elif int(x) == 7: #Get median house price
        POSTCODES = get_postcode()
        result = new_tool.get_median_house_price_estimate(POSTCODES,method=methods['medianPrice'])
        
    elif int(x) == 8: #Estimate total property value
        postal_data = get_postcode()
        result = new_tool.get_total_value(postal_data)
        
    elif int(x) == 9: #Estimate total property value
        POSTCODES = get_postcode()
        result = new_tool.get_annual_flood_risk(POSTCODES)
        
    elif int(x)==99: #Help
        filename = 'file://wsl.localhost/Ubuntu/'+os.getcwd()+'/' + 'html/index.html'
        webbrowser.open_new_tab(filename)  
        result = 'Help open in a new tab'
    else:
        result ='You put the wrong option\nPlease choose 1 from the option'
        
    print ('\033[1m' + "\n",result, "\n")
    print ("______________________________________________________________________\n")
    UserInterface()
```
